### Exercise 1.1
Write a bubble sort function that includes a flag that will allow the function to terminate when there is nothing left to sort.

```python
def bubble_sort(items):
    """
    The buble sort arranges items in a list, and includes a 
    function that terminates when there is nothing left to sort
    """
    while True:
        swapped = False
        for i in range(len(items)):
            for j in range(len(items)-1-i):
                if items[j] > items[j+1]:
                    items[j], items[j+1] = items[j+1], items[j]
                    swapped = True
        if not swapped:
            break
                
    return items
    
```

### Exercise 2.1
Write a bubble sort function that includes a flag that will allow the function to terminate when there is nothing left to sort.

```python
def bubble_sort(items):
    n = len(items)
    while True:
        swapped = False
        for i in range(1, n):
            if items[i-1] > items[i]:
                items[i-1], items[i] = items[i], items[i-1]
                swapped = True
        if not swapped:
            break
    return items
```
