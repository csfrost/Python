### Exercise 1
Convert the mathematical function $f(n) = a^n$ into a recursive function.

```python
def exponential(a, n):
    """ This function raises the variable a, n number of times"""
    if n == 0:
        return 1
    if n > 0:
        return a * exponential(a, n-1)
    else:
        return 1/exponential(a, -n)
```

```python
exponential(2, 3)
```

```python
exponential (2, -3)
```

### Exercise 2
Write a function that uses recursion to calculate the $n$-th number in the fibonacci sequence (0, 1, 1, 2, 3, 5, 8, 13, 21...).

```python
def fibonacci(n):
    """ 
    This function calculates the nth term 
    of a fibonacci sequence.
    """
    if n >= 0:
        if n <= 1:
            return n
        return fibonacci(n-2) + fibonacci(n-1)
    else:
        return (-1)**(n+1) * fibonacci(-n)
```

```python
fibonacci(-6)
```

```python
fibonacci(6)
```

### Exercise 3
Write a recursive function called `cumulate`. The function calculates the sum of all positive integers up to a given number `n`.

```python
def cumulate(n):
    """ 
    This function calculates the sum of all 
    positive integers up to a number n
    """
    if n >= 0:
        if n <= 1:
            return n
        return n + cumulate(n-1)
    else:
        return "Please input a positive digit"
```

```python
cumulate(5)
```

```python
cumulate(-10)
```

### Exercise 4
Write a typical `merge sort` algorithm

```python
def merge_sort(arr):
    """ This function arranges an array 
    in their correct order.
    """
    if len(arr) > 1:
        mid = len(arr) // 2
        l = arr[:mid]
        r = arr[mid:]

        l = merge_sort(l)
        r = merge_sort (r)

        merged = []
        while len(l) > 0 and len(r) > 0:
            if l[0] <= r[0]:
                merged.append(l.pop(0))
            else:
                merged.append(r.pop(0))
        
        merged = merged + l + r
        
        return merged
    
    return arr
    
```

```python
arr = [5, 4, 9, 11, 17, 1, 3, 15]
merge_sort(arr)
```
