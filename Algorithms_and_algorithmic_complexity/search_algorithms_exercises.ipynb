### Exercise 1
write a basic linear search function

```python
def linear_search(list, target):
    """
    search for the target within the list and returns the index of 
    the target if found. If target not found, returns "not found".

    Parameters
    -------
    list: contains a numbers of items
    target: the specific item being search for in the list
    """
    for items in range(len(list)):
        if target == list[items]:
            return items
        else:
            "Not found"
```

Given the below data, use the `linear_search` function to search for the target value.

`list = [1, 3, 5, 7, 9]`

`target = 5`

```python
linear_search([1,3,5,7,9], 5)
```

### Exercise 2
Write a linear search function that counts the number of occurencies of a searched term.

```python
def linear_occurency_search(list, target):
    """
    counts the number of times the target appears in the list
    """
    target_count = 0
    for items in list:
        if items == target:
            target_count +=1 
    return target_count
```

Count the number of times the target value appears in the list below.

`list = [1, 3, 3, 5, 7, 9, 3]`

`target = 3`

```python
linear_occurency_search([1, 3, 3, 5, 7, 9, 3], 3)
```

### Exercise 3
Write a basic recursive implementation of a binary search function

```python
def binary_search_recur(my_list, target_value):
    """
    searches the list to find the target value
    """
    mid = len(my_list)//2
    if len(my_list) == 1:
        return 0 if my_list[0] == target_value else "None"
    
    elif my_list[mid] == target_value:
        return mid
    
    elif target_value < my_list[mid]:
        return binary_search_recur(my_list[:mid], target_value)
    
    else:
        return mid + binary_search_recur(my_list[mid:], target_value)
```

Use the `binary_search_recur` function to search for the target value.

`my_list = [1, 3, 5, 7, 9, 11, 13, 15, 17]`

`target_value = 13`

```python
binary_search_recur([1, 3, 5, 7, 9, 11, 13, 15, 17], 13)
```

### Exercise 4
Write a basic iterative implementation of a binary search function

```python
def binary_search_iter(my_list, target_value):
    """
    Search that iterates over a list to find the target value
    """
    start_value = 0
    end_value = len(my_list) - 1

    while start_value <= end_value:
        mid = (start_value + end_value)//2
    
        if my_list[mid] == target_value:
            return mid
        
        if target_value < my_list[mid]:
            end_value = mid - 1
        
        else:
            start_value = mid + 1
    
    return False
```

Use the `binary_search_iter` function to search for the target value.

`my_list = [1, 3, 5, 7, 9, 11, 13, 15, 17]`

`target_value = 13`

```python
binary_search_iter([1, 3, 5, 7, 9, 11, 13, 15, 17], 13)
```
