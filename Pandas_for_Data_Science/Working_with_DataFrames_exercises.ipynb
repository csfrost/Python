```python
import pandas as pd
import numpy as np
```

We are Using the dataset named `Animals.csv` that stores information about a variety of animal species.

dataset: https://raw.githubusercontent.com/Explore-AI/Public-Data/master/Data/Python/Animals.csv

```python
animals_data = pd.read_csv('https://raw.githubusercontent.com/Explore-AI/Public-Data/master/Data/Python/Animals.csv', index_col= 'Species')

animals_data.head()
```

#### Exercise 1
Filter the dataset to show only species that have an `Average Lifespan (Years)` greater than 20 and are classified as `Vulnerable` in the `Conservation Status`.

```python
#Filtering the data based on the two conditions
longlived_and_vulnerable = animals_data[(animals_data['Average Lifespan (Years)'] > 20) & (animals_data['Conservation Status'] == 'Vulnerable')]

# Displaying the filtered data
longlived_and_vulnerable
```

#### Exercise 2
**a.)** Create a new column `Lifespan in Months` by converting the `Average Lifespan (Years)` to months

```python
def years_to_months(x):
    """ This function multiplies the value of x by 12"""
    return x * 12

# creating the Lifespan in Months column by applying the function 
# to the Average Lifespan (Years) column
animals_data['Lifespan in Months'] = years_to_months(animals_data['Average Lifespan (Years)'])

# Displaying the data
animals_data.head()
```

**b.)** Delete the `Lifespan in Months` from the DataFrame.

```python
# Using the .drop function to delete the Lifespan in Months column
animals_data = animals_data.drop('Lifespan in Months', axis=1, inplace=True)

animals_data.head()
```

#### Exercise 3
**a.)** Filter the DataFrame to include only species with `Average Lifespan (Years)` **greater than 50 years**. Store the result in a new DataFrame `Long_lived_species`.

```python
# Creating a dataframe for older than 50 years species
Long_lived_species = animals_data[animals_data['Average Lifespan (Years)'] > 50]

# Displaying the data
Long_lived_species
```

**b.)** Using the `apply` and `lambda` functions, convert the `Average Lifespan (Years)` of all species into float data type in `Long_lived_species`.

```python
# Converting the datatype in the Average Lifespan (Years) column to float
# Using lambda and apply functions
Long_lived_species['Average Lifespan (Years)'] = Long_lived_species['Average Lifespan (Years)'].apply(lambda x: float(x))

# Displaying the new datatype
Long_lived_species.info()
```

#### Exercise 4
List all columns for species that have the string "`Endangered`" as part of their conservation status.

```python
# Selecting columns with endangered species
Endangered_species = animals_data[animals_data['Conservation Status'] == 'Endangered']

# Displaying the results
Endangered_species
```

#### Exercise 5
list the mean `Average Lifespan (Years)` of species for each `Habitat` in descending order.

```python
# Calculating the mean Average Lifespan (Years) for each habitat in descending order

Mean_lifespan_per_habitat = animals_data.groupby('Habitat')['Average Lifespan (Years)'].mean().sort_values(ascending=False)

# Displaying the results
Mean_lifespan_per_habitat
```

#### Exercise 6
Create a column `Threat Level` that categorises species into `High`, `Medium`,  or `Low` based on their `Conservation Status`:

- **High** for `Critically Endangered` and `Endangered`
- **Medium** for `Vulnerable` and `Near Threatened`
- **Low** for `Least Concern`

```python
def threat_status(status):
    """Classifies the status into different threat level"""
    if status in ['Critically Endangered', 'Endangered']:
        return 'High'
    elif status in ['Vulnerable', 'Near Threatened']:
        return 'Medium'
    else:
        return 'Low'

# Applying the function on the Conservation status column      
animals_data['Threat Level'] = animals_data['Conservation Status'].apply(threat_status)

# Displaying the result
animals_data
```
