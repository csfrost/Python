#### Exercise 1
Create a Pandas DataFrame from a dictionary. The dictionary keys should be `Crop`, `Yield_per_Acre`, and `Country`, and the values should be lists, as shown in the data below:

- Crop: `['Wheat', 'Corn', 'Rice', 'Soybean']`
- Yields per acre (in tons): `[3, 4.5, 5, 2.8]`
- Country: `['USA', 'China', 'India', 'Brazil']`

```python
import pandas as pd
import numpy as np
```

```python
# Creating the columns and values using the data
agric_data = {
    'Crop': ['Wheat', 'Corn', 'Rice', 'Soybean'],
    'Yields_per_Acre': [3, 4.5, 5, 2.8],
    'Country': ['USA', 'China', 'India', 'Brasil']
}

# Creating the dataframe for the agriculture data
agric_df = pd.DataFrame(data=agric_data)

# Printing the DataFrame
agric_df
```

#### Exercise 3
- Using the Numpy library, create a 3x2 array named `tree_array` to store informations for different tree species as follows:

    [[30, 25], [60, 30], [25, 20]].
- Use the array to create a DataFrame with the following columns: `Max_Height_m`, and `Growth_rate_cm_per_year`, and the following index: `Maple`, `Pine` and `Birch`.

```python
# Creating the numpy 3x2 array
tree_array = np.array([[30, 25], [60, 30], [25, 20]])

# Defining the column names
columns = ['Max_Height_m', 'Growth_rate_cm_per_year']

# Defining the index
index = ['Maple', 'Pine', 'Birch']

# Creating the DataFrame
tree_df = pd.DataFrame(data=tree_array, columns=columns, index=index)

#Printing the DataFrame
print(tree_df)
```

#### Exercise 4
Using the DataFrame in Exercise 3, display the Max_Height_m and Growth_rate_cm_per_year columns for the `Pine` tree.

```python
# Displaying the Maximum Height and Growth rate for Pine
Pine_data = tree_df.loc['Pine']

# Printing the pine data
print(Pine_data)
```
