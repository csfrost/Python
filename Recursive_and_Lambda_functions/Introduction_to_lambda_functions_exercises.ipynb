### Exercise 1
Create a function called root to calculate the `y`th root of a number `x`.

```python
def root(x, y):
    if y == 0:
        return "Indeterminate"
    return x ** (1/y)
```

```python
root(8, 2)
```

Create an equivalent lambda function

```python
rooty = lambda x, y: "Indeterminate" if y == 0 else x ** (1/y)
```

```python
rooty(8, 2)
```

### Exercise 2
Write a lambda function named `sort`. This function should take a single argument: a list of numbers and return a new list with the sorted numbers in ascending order.

```python
sort = lambda x: sorted(x)
```

```python
sort([6, 2, 3, 5, 9, 1])
```

### Exercise 3
Write a lambda function named `filterd_list` to print a `list` of integers that are greater than 9.

```python
list1 = [2, 6, 8, 10, 11, 4, 12, 7, 13, 17, 0, 3, 21]

filtered_list = list(filter(lambda x: x > 9, list1))
print(filtered_list)
```

### Exercise 4
Write a lambda function named `mapped_list` to output the remainder after dividing each number in `list2` by 2.

```python
list2 = [2, 6, 8, 10, 11, 4, 12, 7, 13, 17, 0, 3, 21]

mapped_list = list(map(lambda x: x % 2, list2))
print(mapped_list)
```
