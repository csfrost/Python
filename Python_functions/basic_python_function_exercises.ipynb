### Exercise 1
##### Given a list of deforested land in hectares over five years, calculate the total carbon emission over the years using built-in functions. Total carbon emissions are calculated by adding the hectares of deforested land and multiplying the total by a given carbon emissions factor of 40.

```python
# Deforested land for each year
deforested_land = [10, 15, 8, 12, 20]

# carbon emission factor (tons of CO2 per hectare)
carbon_emission_factor = 40

#Calculate the toal carbon emission using built in functions
total_deforested_land = sum(deforested_land)
total_carbon_emission = total_deforested_land * carbon_emission_factor

print(f"The total deforested land is: {total_deforested_land} hectares")
print(f"The total carbon emission is: {total_carbon_emission} tons of CO2")
```

### Exercise 2
##### Create a function named acres_to_hectares that accepts a parameter of the area of a forest in acres and returns the corresponding area in hectares by multiplying the acres by a coversion rate of 0.404686

```python
def acres_to_hectares(area_in_acres):
    """
    This function converts the area_in_acres to area_in_hectares
    by multiplying the area in acres by 0.404686

    Parameter
    -------
    area_in_acres: This is the overall land mass measures in acres
    0.404686: A constant

    Return
    -------
    A float is returned which represents the total land mass
    in hectares
    
    """
    area_in_hectares = area_in_acres * 0.404686
    return area_in_hectares
```

```python
print(acres_to_hectares(50))
```

### Exercise 3
##### Using the data representing the deforested land in hectares over the years, calculate and print the average deforested land using the `sum` and `len` built-in functions

```python
# Deforest land for each year
deforested_land = [10, 15, 8, 12, 20]

# Calculating for the average deforested land
average_deforested_land = sum(deforested_land)/len(deforested_land)

print(f"The average deforested land is: {average_deforested_land} hectares")
```
